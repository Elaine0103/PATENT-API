```python
import pandas as pd
from pandas.io.json import json_normalize
import requests
import json
from urllib3.exceptions import InsecureRequestWarning
from urllib3 import disable_warnings
```

```python
df = pd.read_csv("/Users/elaine/Desktop/side_project/stock/stock.csv",nrows =8,usecols=["main_stock_symbol","sub_stock_symbol","Company_E"])
df
```

```python
Company_E = df["Company_E"].to_list()
Company_E
```

```python
url = "https://api.patentsview.org/patents/query"
json_arr = []
for i in Company_E:

    data = {"q":{ "_and":[{"assignee_organization":i},{"_gte":{"patent_date":"2012-01-01"}}]}, "f":["patent_number","assignee_organization","patent_abstract","patent_date","inventor_first_name","inventor_last_name","cpc_subgroup_id","app_date"],"o":{"page":1,"per_page":10}}
    r = requests.post(url,  json = data)

    ans = json.loads(r.text)
    json_arr.append(ans)
    with open("/Users/elaine/Desktop/side_project/stock/data.json", "w") as outfile:
        json.dump(json_arr, outfile)
```

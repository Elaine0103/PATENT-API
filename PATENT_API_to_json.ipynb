```python
url = "https://api.patentsview.org/patents/query"
json_arr = []
for i in Company_E:

    data = {"q":{ "_and":[{"assignee_organization":i},{"_gte":{"app_date":"2012-01-01"}}]}, "f":["patent_number","assignee_organization","patent_abstract","patent_date","inventor_first_name","inventor_last_name","cpc_subgroup_id","app_date"],"o":{"page":1,"per_page":10}}
    r = requests.post(url,  json = data)

    ans = json.loads(r.text)
    json_arr.append(ans)
    with open("/Users/elaine/Desktop/side_project/stock/data_test.json", "w") as outfile:
        json.dump(json_arr, outfile)
```

```python
df = pd.read_json("/Users/elaine/Desktop/side_project/stock/data_test.json")
df
result = pd.DataFrame.from_records(df.patents.apply(lambda x: x[0]))
```

```python
cpcs = []
for cpc_subgroup_id in result["cpcs"]:   
    
    cpc_set = []
    for sub_cpc in cpc_subgroup_id:
        cpc = sub_cpc["cpc_subgroup_id"]
        ch = "/"
        cpc = cpc.split(ch,1)[0] 
        cpc_set.append(cpc)    
    cpc_set = list(set(cpc_set))
    cpcs.append(cpc_set)

cpcs_list = []
for cpc_str in cpcs:
    cpc_join = ",".join(cpc_str)
    cpcs_list.append(cpc_join)

print(cpcs_list)
```

```python
result["cpcs"] = cpcs_list
result
```

```python
app_dates = []
for applications in result["applications"]:
    
    dates = list(map(lambda d: d['app_date'], applications))
    #print(app_dates)
    for date in dates:
        #print(date)
        app_dates.append(date)
print(app_dates)
```

```python
result["applications"] = app_dates
result
```

```python
assignees = []
for orgs in result["assignees"]:
    orgs_list = []
    for org in orgs:
        assignee_organization = org["assignee_organization"]
        orgs_list.append(assignee_organization)
    #print(orgs_list)
    assignees.append(orgs_list)
#print(assignees)
assignees_list = []
for assi in assignees:
    assi_join = ";".join(assi)
    assignees_list.append(assi_join)
print(assignees_list)
```

```python
result["assignees"] = assignees_list
result
```
